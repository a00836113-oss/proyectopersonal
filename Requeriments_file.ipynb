En esta celda, se genera el arxchivo requirements.txt que contendrá todas las librerías que utiliza tu app para que se pueda guardar en Github

```python
packages = ["pandas", "numpy", "matplotlib", "scikit-learn", "plotly", "streamlit"]
with open("requirements.txt", "w") as f:
    for pkg in packages:
        version = !pip show {pkg} | grep Version
        version = version[0].split(":")[1].strip()
        f.write(f"{pkg}=={version}\n")
```
